```python
#importar líbrerias
import pandas as pd
import plotly.express as px
```

```python
#Cargar archivo
car_data = pd.read_csv('../vehicles_us.csv') # leer los datos
```

```python
#Hacer una copia del archivo original
car_copy=car_data.copy()
```

```python
#Separar modelo y marca de la columna "model"
car_copy["model"]=car_copy["model"].str.split()
```

```python
#Crear nueva columna para marca
#Usar una funcion lambda para obtener solo la marca de los vehículos y hacer mejores graficos sin tanta información
car_copy["brand"] = car_copy["model"].apply(lambda x: x.pop(0) if len(x) > 0 else x)
```

```python
#verifcar el cambio
car_copy["brand"].unique()
```

```python
#Histograma
fig_1 = px.histogram(car_copy, x="brand")
fig_1.show()
```

```python
#Grafíco de dispersión
fig_2= px.scatter(car_copy, x="brand", y="price", color="condition")
fig_2.show()
```
